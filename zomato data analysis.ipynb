## Zomato Data Analysis

## Importing the libraries

```python
#Loading the dataset
import pandas as pd
zom = pd.read_csv("C:/Users/bhavi/Downloads/zomato.csv")
zom.head()
```

```python
zom.head(10)
```

## convert the datatype of column - rate

```python
def handleRate(value):
    value = str(value).split('/')[0]
    if value == 'NEW' or value == '-':
        return None
    return float(value)
zom['rate'] = zom['rate'].apply(handleRate)
zom.head()
```

## What type of restaurants do majority of the customers order from?

```python
#calculating online order restaurants
O_rest = zom[zom['online_order'] == 'Yes']
O_rest
```

```python
#grouping by restaurant types
rest_type = O_rest['rest_type'].value_counts()
rest_type
```

```python
#visualizing the above data
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,5))
sns.barplot(x = rest_type.index, y = rest_type.values)
plt.title('Most Popular Restaurant Types for Online Orders')
plt.xlabel('Restaurant Type')
plt.ylabel('Number of Orders')
plt.xticks(rotation=90)
plt.show()
```

## How many votes has each type of restaurant recieved by customers?

```python
votes_by_rest = zom.groupby(['rest_type']).agg({
'votes' : 'sum'
}).reset_index();
```

```python
votes_by_rest
```

## Top 20 Restaurant types by customer votes

```python
top_20_rest_by_votes = votes_by_rest.sort_values(by = 'votes',ascending = False).head(20)
```

```python
top_20_rest_by_votes
```

```python
import squarify
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))

# Create the treemap
squarify.plot(sizes=top_20_rest_by_votes['votes'], 
              label=top_20_rest_by_votes['rest_type'], 
              alpha=0.8, 
              color=plt.cm.Paired.colors)

# Add a title
plt.title('Top 20 Restaurant Types by Votes')

# Remove axis for better presentation
plt.axis('off')

# Show the plot
plt.show()
```

## What are the ratings that majority of Restaurants has recieved

```python
ratings_count_by_rest = zom.groupby('rate').size().reset_index(name = 'count')
ratings_count_by_rest = ratings_count_by_rest.sort_values(by = 'count',ascending=False)
ratings_count_by_rest
```

```python
import matplotlib.pyplot as plt

# Increase figure size
plt.figure(figsize=(8,8))
plt.barh(ratings_count_by_rest['rate'], ratings_count_by_rest['count'], color='skyblue')
plt.xlabel('Number of Restaurants')
plt.ylabel('Rating')
plt.title('Distribution of Ratings Among Restaurants')
plt.show()
```

```python
online_orders = zom[zom['online_order'] == 'Yes']
online_orders

online_orders['approx_cost(for two people)'] = pd.to_numeric(online_orders['approx_cost(for two people)'], errors='coerce')
```

## Average cost for 2 people according to Restaurant

```python
avg_cost = online_orders.groupby(['name']).agg({
'approx_cost(for two people)' : 'mean'
}).reset_index()
avg_cost

```

```python
max_ratings = zom.groupby('online_order').agg({
    'rate': 'max'
}).reset_index()
max_ratings
```

## Which type of restaurants recieved more offline orders?

```python
offline_orders = zom[zom['online_order'] == 'No']
offline_orders_count = offline_orders.groupby('rest_type').size().reset_index(name='offline_orders_count')
offline_orders_count
```

```python
most_offline_orders_type = offline_orders_count.loc[offline_orders_count['offline_orders_count'].idxmax()]
most_offline_orders_type
```

```python
plt.figure(figsize=(12,20))
plt.barh(offline_orders_count['rest_type'], offline_orders_count['offline_orders_count'], color='pink')
plt.xlabel('Number of Offline Orders')
plt.ylabel('Restaurant Type')
plt.title('Number of Offline Orders by Restaurant Type')
```
